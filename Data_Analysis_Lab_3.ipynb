**1--  Run a simple bivariate regression, and interpret your results. (Did the results fit your expectations? Why? Why not?)**


Dependent Variable: The current player's value (value).

```python
df[['value']].describe()
```

Independent Variable: he player's age in years (age).



```python
df[['age']].describe()
```

I expect that there is a relationship where players with higher age tend to have less price in million.

```python
age_price = smf.ols(formula = 'value~age', data = df).fit()
print (age_price.summary())
```

Coefficient for age is negative 307900, which indicates that for each additional age increase suggests lower player's value, such that the player's value decreases by approximately 307900. The model is statistically significant with p < 0.001. The R-squared of 0.007 indicats a weak association.


**2--  Add an additional variable that might mediate or partly "explain" the initial association from that simple regression above -- and explain your results. Did it work out? Yes? No?**


The average minutes (minutes_90s) is added. This will allow us to see if the average minutes would affect the players' value.

```python
df[['minutes_90s']].describe()
```

How do my two independent variables (two measure of corrpution) correlate?  They do have some correlate(+0.18).

```python
df_filtered = df[['age', 'minutes_90s']].dropna()
stats.pearsonr(df_filtered['age'], df_filtered['minutes_90s'])
```

So this regression will involve the average minutes (minutes_90s) is added. This will allow us to see if average minutes can explain the association between player's value and their ages.

```python
age_minutes_price = smf.ols(formula = 'value ~ age + minutes_90s', data = df).fit()
print (age_minutes_price.summary())
```

Average minutes has a coefficient of 556900, indicating that a 1 minute increase is associated with an increase of 556900 in player's value.

Player's age has a coefficient of -549100, suggesting that as age increase by 1 year, player's value decrease by approximately 549100.

The adjusted R-squared is 0.134, indicating that about 13.4% of the variance in player's value is explained by the two variables combined.
Both corruption variables are statistically significance, with p-values < 0.01.

This result seems weak since only 13.4% of vairance in player's value is explained. However, after we add variable average minutes, our R-square increases. Thus, we do need add more relevant predictors like reputation or market factors to the measurement.

**3-- More on extreme combinations.  Find the top 5 entities that are ranked at the top on one variable and ranked at the bottom on another variable.  Interpret your results.**

I am going to look to players who has a very high player's value, but at the same time have a high age and low average minutes.

```python
df['player_age_rank'] = df['age'].rank(ascending=False)
df['player_minutes_rank'] = df['minutes_90s'].rank(ascending=True)

extreme_comb = df[['player', 'player_age_rank', 'player_minutes_rank']].copy()
extreme_comb['rank_difference'] = extreme_comb['player_age_rank'] - extreme_comb['player_minutes_rank']
extreme_sorted = extreme_comb.sort_values(by='rank_difference', ascending=False).head(5)

extreme_sorted
```

These results highlight a significant difference in these players between the average minutes and age. This could suggest that older players may also have the possibility to have high average minutes.
