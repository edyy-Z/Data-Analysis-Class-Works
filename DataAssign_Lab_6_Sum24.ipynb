1. Run a multiple linear probability model (have at least 2 Xs in the model). Tell me how you think your independent variables will affect your dependent variable. Interpret your results. Were your expectations correct? Why or why not?

Car price is (price) the dependent variable. The data range from 4950 to 74600 with 4950 being cheap car and 74600 being expensive car. This has been recoded to binary so cheap car (4950-15000) is 0 and good economy (15000-74600) is 1. I believe that car price should be associated with a higher gear numbers (Gears), horsepower (hp_KW) , and a lower drived distance in km (km).

```python
sub = df.dropna(subset = ['price', 'Gears', 'hp_kW', 'km'])
```

```python
conditions = [(sub['price'] < 15000) , (sub['price'] > 15000)]
choices = [0, 1]
sub['good_car'] = np.select(conditions, choices, default=np.nan)
```

```python
pd.crosstab(index=sub["good_car"], columns="count")
```

```python
lm1 = smf.ols(formula = 'good_car ~ Gears + hp_kW + km', data = sub).fit()
print (lm1.summary())
```

Number of gears (Gears) is positively associated with the price of a car, and the association is statistically significant.

Horsepower (hp_KW) has a very small coefficient and is not statistically significant.

Drived distance (km) is negatively associated with the price, and the association is statistically significant.

2. Run a multiple (binary) logistic model. (It can be the same as the above LPM or a new model.) If it is a new model, tell me how you think your independent variables will affect your dependent variable. Interpret your results in the logit scale. Were your expectations correct? Why or why not?

```python
logit1 = sm.formula.logit(formula = 'good_car ~ Gears + hp_kW + km', data = sub).fit()
print (logit1.summary())
```

Logit Model shows the same results. Gears is positively associated with price of cars, and statistically significant. Hp_KW coefficient is very small and not statistically significant. Drived distance is negatively associated with the price, and the association is statistically significant.

4. Get predicted probabilities from your logit model in Question 2 for some constellations of X values and interpret the results.

```python
def logit2prob (logit):
    odds = np.exp(logit)
    prob = odds / (1 + odds)
    return(prob);

intercept = logit1.params.Intercept
b_Gears = logit1.params.Gears
b_hp_kW = logit1.params.hp_kW
b_km = logit1.params.km
```

Gears = 7, horsepower = 100, drived distacne = 1000

```python
logits_exh = intercept + (6 *b_Gears ) + (100 * b_hp_kW) + (1000 * b_km)
logit2prob(logits_exh)
```
